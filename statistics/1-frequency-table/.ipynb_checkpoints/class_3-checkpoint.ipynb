```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
data_path = Path('../../datasets/student_mental_health.csv')
df = pd.read_csv(data_path)
df.info()
```

# Graphical Representations of Frequency Tables

Depending on the **variable type** (categorical, discrete, countinous), there are different ways of representing it.

## Categorical variables
(gender, course, state, ...)

### Bar chart
Each category is a bar, and the height of it is the frequency (absolute or relative)

```python
abs_course_freq = df['What is your course?'].value_counts().reset_index().iloc[0:10]
abs_course_freq.columns = ['Course', 'Frequency']
sns.barplot(abs_course_freq, x='Frequency', y='Course')
plt.ylabel('')
plt.title('Top 10 Most Frequent Courses')
```

### Pie plot
Used to show the proportion of each category in a circle **(not recommended)**

```python
rel_course_freq = df['What is your course?'].value_counts(normalize=True).iloc[0:10]
rel_course_freq.plot(kind='pie', autopct='%1.2f%%', ylabel='')
```

## Discrete variables
1, 2, 3, ...
(number of childer, cars, ...)

### Barplot
We can use barplots for discrete numbers, but only if there are only a few of them (not too many bars)


```python
s = pd.DataFrame({
    'x' : [3, 4, 2, 3, 5, 6, 7, 4, 4, 3, 7]
}).value_counts().reset_index()
sns.barplot(s, x='x', y='count')
```

## Continous data
(Real numbers, measurements, weight, height, speed, income, temperature, ...)

This is where **class intervals** and **histograms** come in.

### Histogram
It is the visual representation of the **frequency distribution** of a countinous variable.

Instead of showing individual data points, we group data into classes(bins), and we can see the shape of our data as a result.

```python
values = np.random.normal(100, 15, 50)
a = pd.DataFrame({
    'x': values
})
sns.histplot(a, x='x', bins='fd')
```

**Lets correct the bin edges on the x ticks markers with numpy**

```python
bin_edges = np.histogram_bin_edges(a['x'], bins='fd')
sns.histplot(a, x='x', bins=bin_edges).set_xticks(bin_edges)
```
