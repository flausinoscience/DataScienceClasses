# Measures of Central Tendency
## What they represent?
- They are used to summarize the *center* or *typical value* of a dataset
- The point around which data tend to cluster
- Essential to describe "avarege behavior" of data before doing some **statistic modeling**

## Mean
Represents the *balance point* of the data. It is sensitive to **outliers**.
$$
\LARGE \bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}
$$

The mean is ideal when data is **symmetric** and without extreme outliers

```python
import numpy as np
```

```python
data = [5, 7, 9, 10, 12]
x_bar = np.mean(data)
x_bar
```

## Median (Middle Value)
Is the middle value when data is **ordered**

Half of observations lie below and half above the median.

**Not influenced** by outliers.

### Steps to calcule
- order the data
- if *n* is odd: get the middle value
- if *n* is even: get the **mean** of the two middle values

The median gives the typical value when data are **skewed**.

```python
data_with_outlier = [5, 7, 9, 10, 12, 12349, 3333333]
median = np.median(data_with_outlier)
median
```

## Mode (Most Frequent Observation)
It is the value that occurs most frequently in a dataset. 

Used a lot with **categorical data**.

Examples: (most commom city, color, product type)

```python
from statistics import mode, multimode
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4]
print(mode(data))
print(multimode(data))
```

## Visualizing the differences

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
s = pd.Series([1, 2, 3, 4, 5, 6,7, 7, 20])
sns.histplot(s, bins=8, kde=True, alpha=.2)
plt.axvline(s.mean(), linestyle='--', color='red', label=f'Mean = {s.mean():.2f}')
plt.axvline(s.median(), linestyle='--',color='green', label=f'Median = {s.median()}')
plt.axvline(mode(s), linestyle='--', color='purple', label=f'Mode = {mode(s)}')
plt.legend()
```
